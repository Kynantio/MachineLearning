2. Lakukan proses encoding pada kolom "diagnosis".
3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.
4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
import pandas as pd

data = '/content/wbc.csv' # path dataset
dft = pd.read_csv(data) # load dataset\
dft.head()
```

```python
dft.info()
```

```python
dft = dft[['id', 'diagnosis', 'radius_mean', 'radius_se', 'radius_worst']]
dft.head()
```

```python
le = LabelEncoder()
dft['diagnosis'] = le.fit_transform(dft['diagnosis']) # proses encoding
```

```python
dft.head()
```

```python
std = StandardScaler()
dft['radius_mean'] = std.fit_transform(dft[['radius_mean']])
dft['radius_se'] = std.fit_transform(dft[['radius_se']])
dft['radius_worst'] = std.fit_transform(dft[['radius_worst']])
```

```python
dft.head()
```

```python
# Split Data Stratified
from sklearn.model_selection import train_test_split

dft_train, dft_test, = train_test_split(dft, test_size=0.2, stratify=dft['diagnosis'], random_state=0)

print(f'Jumlah label data asli:\n{dft.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{dft_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{dft_test.diagnosis.value_counts()}')
```
